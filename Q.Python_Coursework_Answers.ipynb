### Question 7
Analyze the life expectancy dataset (which can be loaded by running the cell below).

Calculate the number of unique countries in the dataset.

Calculate the country and year with the highest life expectancy.

Calculate the year with the highest total expenditure.

Provide your code using pandas.


````python
import seaborn as sns
life_expectancy = sns.load_dataset('healthexp')

life_expectancy.head()
````

````python
# Calculate number of unique countries
unique_countries = set(life_expectancy.iloc[:, 1]) # Turns column to a set, only counting unique contries
num_unique_countries = (len(unique_countries)) # Calculates length of the set
print(f"Number of unique Countries: {num_unique_countries}")

# Calculate the country and year with the highest life expectancy.
highestLE = (max(life_expectancy.iloc[:, 3])) # Return highest life expectancy
country_highestLE = life_expectancy[life_expectancy.iloc[:,3] == highestLE] # Find row containing highest life expectancy
print(f"Country with highest life expectancy: {country_highestLE.iloc[0,1]}, Year: {country_highestLE.iloc[0,0]} Life Expectancy: {highestLE}") # Print sections of that row

# Calculate the year with the highest total expenditure
total_expenditure_by_year = life_expectancy.groupby('Year')['Spending_USD'].sum() # Sum total expenditure by year
highest_expenditure = max(total_expenditure_by_year) # Find the maximum total expenditure
year_highest_expenditure = total_expenditure_by_year[total_expenditure_by_year == highest_expenditure].index[0] # Find the corresponding year
print(f"Year with the highest total expenditure: {year_highest_expenditure}, Total Expenditure: {highest_expenditure}")
````

### Question 8
Using the diamonds dataset and the plotting code of your choice, plot the following from the diamonds dataset.

1. Carat vs Price in a scatter plot.
2. A histogram of prices.
3. A boxplot of cut vs price.

The code can be loaded from seaborn in the cell below.

````python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
````

````python
# Carat vs Price in a scatter plot.
plt.scatter(diamonds['carat'], diamonds['price'], color='blue', s=0.5)
plt.xlabel('Carat')
plt.ylabel('Price ($USD)')
plt.title('Carat vs Price')
plt.show()

# A histogram of prices
plt.hist(diamonds['price'], bins=25, color='green')
plt.xlabel('Price ($USD)')
plt.ylabel('Frequency')
plt.title('Histogram of Diamond Prices')
plt.show()

# A boxplot of cut vs price
sns.boxplot(x='cut', y='price', data=diamonds)
plt.title('Diamond Cut vs Price')
plt.show()
````

### Question 10
Write a function that will convert carat to grams and find the heaviest diamond.

1 carat = 0.2 grams

````python
def heaviest_diamond(diamonds):
    grams = diamonds.iloc[:,0] * 0.2
    heaviest = (max(grams))

    return heaviest
heaviest = heaviest_diamond(diamonds)
print(f" The weight of the heaviest diamond is {heaviest} grams")
````
